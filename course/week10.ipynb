## Activity: Dungeons & Dragons Character Hierarchy

In this activity, you will create a simple Dungeons & Dragons character class hierarchy using inheritance. Start with a base class called `Character` and extend it to more specific subclasses like `Wizard`, `Rogue`, and `Warrior`. Each subclass can have unique attributes and methods that reflect their specific character class.

```python
# Dungeons & Dragons Character Hierarchy

class Character:
    def __init__(self, name):
        self.name = name

    def show_info(self):
        print(f"Name: {self.name}")

class Wizard(Character):
    def __init__(self, name, spell_power):
        super().__init__(name)
        self.spell_power = spell_power

    def cast_spell(self):
        print(f"{self.name} casts a powerful spell with spell power {self.spell_power}!")

class Rogue(Character):
    def __init__(self, name, agility):
        super().__init__(name)
        self.agility = agility

    def perform_sneak_attack(self):
        print(f"{self.name} performs a sneak attack with agility {self.agility}!")

class Warrior(Character):
    def __init__(self, name, weapon):
        super().__init__(name)
        self.weapon = weapon

    def attack(self):
        print(f"{self.name} attacks with a {self.weapon}!")

# Create character instances
wizard = Wizard("Gandalf", 100)
rogue = Rogue("Sneaky", 80)
warrior = Warrior("Braveheart", "sword")

# Show character info
wizard.show_info()
wizard.cast_spell()

rogue.show_info()
rogue.perform_sneak_attack()

warrior.show_info()
warrior.attack()
```
